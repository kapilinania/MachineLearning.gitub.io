```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
iris
```

```python
# Access the target names
print(iris.target_names)
```

```python
print(iris.feature_names)
```

```python
import pandas as pd
# Create a Pandas DataFrame with all columns
data = pd.DataFrame({
    'sepal length': iris.data[:, 0],
    'sepal width': iris.data[:, 1],
    'petal length': iris.data[:, 2],
    'petal width': iris.data[:, 3],
    'species': iris.target
})
```

```python
data.head(5)
```

```python
#here we are find depeneded variable and independed variable
#here species is depened variable and inpenedede is remain all
```

```python
x = data[['sepal length', 'sepal width', 'petal length', 'petal width']]  #features
y = data['species'] #label
```

```python
x
```

```python
y
```

```python
#we are split dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y, test_size=0.3)  #70% tranining 30 % test
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#here we are import random forest model
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100, criterion='gini')  #n_estimators  how many trees we arae inlclude this    #criterion     entropy or gini 2 tachinque hotee hia split karn k liyt
```

```python

#herer we are train model
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
#    0 =setosa, 1 = versicolor 2. virgonica
x_test
```

```python
clf.score(x_test,y_test )
```

```python
clf.predict([[3,5,4,2]])
```

```python
clf.predict([[3,1,2,2]])
```

```python
# Calculate feature importance
feature_importance = clf.feature_importances_

# Create a Pandas Series with feature names as the index
feature_importance_series = pd.Series(feature_importance, index=iris.feature_names)

# Sort the Series in descending order
feature_importance_series = feature_importance_series.sort_values(ascending=False)

# Display the sorted feature importance
print(feature_importance_series)
```
